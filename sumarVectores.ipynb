```python
import matplotlib.pyplot as plt

# Define los componentes de los vectores
a = [1, 2]  # Vector a
b = [3, 1]  # Vector b

# Calcula la suma de los vectores
suma = [a[0] + b[0], a[1] + b[1]]

# Crea una figura y un eje para el gráfico
fig, ax = plt.subplots()

# Grafica el vector a
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='a')

# Grafica el vector b
ax.quiver(a[0], a[1], b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='b')

# Grafica la suma de los vectores
ax.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1, color='g', label='a + b')

# Establece los límites del gráfico
plt.xlim(0, max(a[0], b[0], suma[0]) + 1)
plt.ylim(0, max(a[1], b[1], suma[1]) + 1)

# Etiqueta los ejes
plt.xlabel('x')
plt.ylabel('y')

# Muestra la leyenda
plt.legend()

# Muestra el gráfico
plt.grid()
plt.show()
```
